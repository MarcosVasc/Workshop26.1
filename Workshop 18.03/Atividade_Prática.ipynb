```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


temperaturas = np.random.uniform(20,35,7)

dias = ["Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sábado", "Domingo"]

df = pd.DataFrame({
    "Dia": dias,
    "Temperatura (°C)": temperaturas
})

df
```

```python
media = df["Temperatura (°C)"].mean()
maxima = df["Temperatura (°C)"].max()
minima = df["Temperatura (°C)"].min()

print(f"\nMédia: {round(media, 1)} °C")
print(f"Máxima: {round(maxima, 1)} °C")
print(f"Mínima: {round(minima, 1)} °C")
```

```python
plt.plot(dias, temperaturas)
plt.title("Temperatura na Semana (Linha Simples)")
plt.xlabel("Dias")
plt.ylabel("Temperatura (°C)")
plt.grid()
plt.show()
```

```python
plt.plot(dias, temperaturas, marker='o')
plt.title("Temperatura com Marcadores")
plt.xlabel("Dias")
plt.ylabel("Temperatura (°C)")
plt.grid()
plt.show()
```

```python
plt.plot(dias, temperaturas, linestyle='--')
plt.title("Temperatura (Linha Tracejada)")
plt.xlabel("Dias")
plt.ylabel("Temperatura (°C)")
plt.grid()
plt.show()
```
